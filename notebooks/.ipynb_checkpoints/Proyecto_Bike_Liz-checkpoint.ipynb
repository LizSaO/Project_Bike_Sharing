**Escalar las varibles continuas**

```python
scaler = MinMaxScaler()
```

**Modelado**

```python
np.random.seed(0)
bike_train, bike_test = train_test_split(bike_new, train_size = 0.80, test_size = 0.20, random_state = 42)
```

```python
bike_test.shape
```

```python
bike_train.shape
```

```python
y_train = bike_train.pop('cnt')
X_train = bike_train

y_test = bike_test.pop('cnt')
X_test = bike_test
```

Seleccionaremos a las mejores 25 variables

```python
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(estimator=lm,n_features_to_select=25)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
#list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
# Modelo de regresion lineal
X_train_rfe = X_train[col]
X_test_rfe = X_test[col]

modelo= LinearRegression()
modelo.fit(X_train_rfe, y_train)
```

```python
#Validar con el conjunto de test
y_pred = modelo.predict(X_test_rfe)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('MSE:', mse)
print('R2:', r2)
```

Modelo final

```python

from sklearn.pipeline import Pipeline
pipe = Pipeline(steps=[('scaler', MinMaxScaler()),
                   ('rfe', RFE(estimator=LinearRegression(),n_features_to_select=25)),
                   ('modelo', LinearRegression())])
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print('MSE:', mse)
print('R2:', r2)
```


Pipeline Final

```python
with open('trained_model-1.0.0.pkl','wb') as f:
  pickle.dump(pipe, f)
```

```python
with open('trained_model-1.0.0.pkl', 'rb') as f:
  model = pickle.load(f)
```

```python
#defici√≥n de la funci√≥n
def predict_cnt(data):

   data_df = pd.DataFrame(data, index=[0])
   for variable in Vard:
    data_df[variable] = data_df[variable].astype('category')
   print("Listo catgoricas")

   data_df = pd.get_dummies(data_df, drop_first=True)
   print("Listo dummies")
   print(data_df.info)

   data_df[Varc] = scaler.fit_transform(data_df[Varc])
   print("Listo escalar")

   prediction = model.predict(data_df)[0]

   return prediction
```

```python
# Validar

data = {'season': 1, 'yr': 1, 'hr': 10, 'mnth': 1, 'holiday': 0, 'weekday': 1,
       'workingday': 1, 'weathersit': 1, 'temp': 0.344146, 'atemp': 0.363628,
       'hum': 0.805825, 'windspeed': 0.160446}

prediction = predict_cnt(data)
print(f"Predicted number of bike rentals: {prediction}")
```
